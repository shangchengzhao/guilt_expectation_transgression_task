```python
import pandas as pd
import random

DVs = [0, 1, 2, 3]  # Initialize DVs list
buttons = {0: ["Greater", "Smaller"], 1: ["Smaller", "Greater"]}  # Define buttons dictionary
conditions = {0: "both_right", 1: "both_wrong", 2: "only_sub_right", 3: "only_sub_wrong"}  # Define conditions dictionary
dotimg = [f"RandomDot{i}.png" for i in range(1, 41)]  # Create list of dot images
repeat = 6  # Define repeat value

# Initialize empty DataFrame for the result table
tb = pd.DataFrame({
    "type_trial": [], "stipic": [], "DV": [],
    "dotlefttext": [], "dotrighttext": [], "picture": [],
    "answer": [], "number": []
})

# Generate experimental trials
for key, value in conditions.items():
    for DV in DVs:
        for button in buttons.values():
            for rep in range(repeat):
                tb = tb.append({
                    'type_trial': key,
                    'stipic': value + ".png",
                    'DV': DV,
                    'dotlefttext': button[0],
                    'dotrighttext': button[1],
                    'picture': random.choice(dotimg),
                    'answer': "No answer",
                    'number': 20 + random.randint(-1, 1)
                }, ignore_index=True)

print(tb)

```

```python
# attention check
repeat = 3
for this_type in [4,5]:
    for bk, button in buttons.items():
        for rep in range(repeat):
            answer = "LeftButton" if ((this_type == 4 and bk == 0) or (this_type == 5 and bk == 1)) else "RightButton"
            number = 5 if this_type==4 else 20 + random.randint(-1, 1)
            tb = tb.append({
                            'type_trial': this_type,
                            'stipic': "both_right.png",
                            'DV': 10,
                            'dotlefttext': button[0],
                            'dotrighttext': button[1],
                            'picture': f"RandomDot{37+this_type}.png",
                            'answer': answer,
                            'number': number
                        }, ignore_index=True)
tb.tail(10)
```

```python
# for each subject, randomly choose 3 trials from each condition and 1 trial from the attention check
# experimental design - practice
import random as random
pracrow = random.sample(range(10), 3)

# exp design - formal procedure
chunks = [range(i, i+12) for i in range(0, 192, 12)]
chunks.append(range(192, 198)) # attention check: trial_type = 4
chunks.append(range(198, 204)) # attention check: trial_type = 5

# Initialize a list to store the selected trials
chooserow = []

# For the both_right condition, randomly choose 1 trials from each DV
for chunk in chunks[:4]:
    choose = random.sample(chunk, 1)
    chooserow.extend(choose)
    print(f"experiment: {chunk}, choose: {choose}")

# Randomly choose items from the first 16 chunks (3 items per chunk)
for chunk in chunks[4:len(chunks)-2]:
    choose = random.sample(chunk, 3)
    print(f"experiment: {chunk}, choose: {choose}")
    chooserow.extend(choose)

# Randomly choose items from the last 2 chunks (1 item per chunk)
for chunk in chunks[len(chunks)-2:]:
    choose = random.sample(chunk, 1)
    chooserow.extend(choose)
    print(f"experiment: {chunk}, choose: {choose}")

tb.loc[chooserow, :]

chooserow = random.sample(chooserow, len(chooserow))
for i in range(0, len(chooserow)-2): # pseudo-randomize the order of the trials to make sure that the same condition does not appear consecutively
    while chooserow[i] //12 % 4 == chooserow[i+1] // 12 % 4 == chooserow[i+2] // 12 % 4:
        tail = chooserow[range(i+2, len(chooserow))]
        newtail = random.sample(tail, len(tail))
        chooserow = chooserow[range(i+2)] + newtail

print(f"chooserow: {chooserow}")
```
